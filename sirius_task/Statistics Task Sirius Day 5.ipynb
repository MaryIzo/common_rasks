Задание 3*.

Известно, что все люди делятся на две группы: тех, кто чистит банан со стороны черенка (т. н. "правильный способ") и тех, кто чистит банан со стороны носика (т. н. "неправильный способ"). Предположим, что в обществе всего 100 человек, из которых 32 чистят банан правильным способом, а 68 - неправильным. Так же у каждого человека есть какое-то количество друзей, и каждый человек, заводя дружбу, может заводить её с человеком который чистит банан так же, либо с тем, который чистит банан по-другому. Посчитано, что люди, чистящие банан правильным способом, всего имеют 120 дружеских связей с людьми, так же чистящими банан правильным способом и 181 связь с людьми, чистящими банан неправильным способом (120 внутригрупповых и 181 межгрупповую связь). Представители группы чистящих банан неправильным способом имеют 326 внутригрупповых связей и 181 межгрупповую (те же самые). Каждая пара человек может дружить, а может не дружить. Так же каждая пара человек может иметь одинаковые взгляды на чистку банана, а может разные. Проверить гипотезу о том, что люди выбирают себе друзей не взирая на то, с какого конца те чистят банан.
